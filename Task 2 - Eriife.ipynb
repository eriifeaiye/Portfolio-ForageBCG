## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

The first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

## Correlation Analysis
Tests to see if variables have a linear relationship between each other

### Consumption patterns vs Customer churn

```python
# corr btw electricity & churn -0.045968
# corr btw gas and churn -0.037957

# test elec, gas and churn
cons_cols = ['cons_12m', 'cons_gas_12m', 'churn']
correlation_cons = client_df[cons_cols]
corr_cons_matrix = correlation_cons.corr()
print(corr_cons_matrix) # 0.488474 customers tend to inc elec cons at the same level as inc gas cons over 12 months
```

### Pricing patterns vs Customer churn

```python
full_df = pd.merge(client_df, price_df, on='id', how='inner')

# Price of energy in peak variations 
energyPrice_cols = ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'churn'] # -0.0061 0.0293 0.0461
corr_energyPrice = full_df[energyPrice_cols]
energyPrice_corr_matrix = corr_energyPrice.corr().round(4)
print(energyPrice_corr_matrix) #changes in energy prices ALONE across peak = little to no impact on churn

#Price of power in peak variations
powerPrice_cols = ['price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix', 'churn'] # 0.0166 0.0469 0.0444
corr_powerPrice = full_df[powerPrice_cols]
powerPrice_corr_matrix = corr_powerPrice.corr().round(4)
print(powerPrice_corr_matrix) # corr btw power prices & churn are weak = little/no direct impact on churn
```

Combined consumption (electricity and gas together) has average impact on churn.
Individually, peak prices have little effect on churn, so I need to look at total prices (net, forecasts, margins, etc)

### Sales channel

```python
channel = client_df[['id', 'channel_sales', 'churn']]
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1).fillna(0)
channel_churn = (channel.div(channel.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

plot_stacked_bars(channel_churn, "Sales channel churn impact", rot_=30, legend_="lower right")
```

Churned customers make up 5 out of the different 8 sales channels. A relatively high percentage (7.6%) comes from a channel labeled 'MISSING' which may be important in modeling later.

### Client contract

```python
contract = client_df[['id', 'has_gas', 'churn']]
# categorical object so treat with stacked bars 
contract = contract.groupby([contract['has_gas'], contract['churn']])['id'].count().unstack(level=1).fillna(0)
contract_churn = (contract.div(contract.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)
contract_churn.index = contract_churn.index.map({'f': 'False', 't': 'True'})


plot_stacked_bars(contract_churn, "Client contract churn impact", legend_="lower right")
```

### Margins

```python
margin = client_df[['id', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'net_margin', 'churn']]

fig, axs = plt.subplots(nrows=3, figsize=(18, 25))

plot_distribution(margin, 'margin_gross_pow_ele', axs[0])
plot_distribution(margin, 'margin_net_pow_ele', axs[1])
plot_distribution(margin, 'net_margin', axs[2])
```

Clearly the margin data is highly positive skewed presenting a very long right-tail towards the higher values of the distribution. The values on the higher and lower end of the distribution are likely to be outliers. To view the outliers in more detail, I will use box plots to visualise the margin data.

### Client history

```python
history = client_df[['id', 'nb_prod_act', 'num_years_antig', 'origin_up', 'churn']]

products = history.groupby([history['nb_prod_act'], history['churn']])['id'].count().unstack(level=1).fillna(0)
products_churn = (products.div(products.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

plot_stacked_bars(products_churn, "Active products churn impact", legend_="lower right")
```

As can be observed, clients with more active products and services are less likely to churn (100% retention). Clients with little subscribed products and services are mor elikely to churn. 
When it comes to feature engineering this may be important in building a model to keep customers from churning by motivating them to have more active products and services.

```python
antiquity = history.groupby([history['num_years_antig'], history['churn']])['id'].count().unstack(level=1).fillna(0)
antiquity_churn = (antiquity.div(antiquity.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

plot_stacked_bars(antiquity_churn, "Client antiquity churn impact", legend_="lower right")
```

As observed, customers with >1 year in their length of time as a client of the company have more churn rates. 

```python
campaign = history.groupby([history['origin_up'], history['churn']])['id'].count().unstack(level=1).fillna(0)
campaign_churn = (campaign.div(campaign.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

plot_stacked_bars(campaign_churn, "Electricity campaign churn impact", rot_=30, legend_="lower right")
```

The different campaigns customers first subscribed through track what first attracted the customer to the company"s electricity services. 
As observed the last two campaigns in the chart still have all the customers retained. However, the other four have churn rates including the `MISSING` campaign which may be useful to explore when modeling.
